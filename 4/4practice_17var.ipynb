```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score as det
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
X = []
Y = []

types = iris.species.values
petal_width = iris.sepal_width.values
sepal_length = iris.petal_length.values

for i in range(1, len(iris), 1):
    if types[i] == 'versicolor':
        X.append(petal_width[i])
        Y.append(sepal_length[i])
```

```python
plt.scatter(X, Y)

plt.xlabel('petal_width')
plt.ylabel('sepal_length')
plt.grid()

plt.show()
```

```python
ind = np.lexsort((X, Y))

X1 = []
Y1 = []

for c in ind:
    X1.append(X[c])
    Y1.append(Y[c])
```

```python
Y2 = []

for x in X1:
    Y2.append(np.poly1d(np.polyfit(X1, Y1, 1))(x))

det(Y1, Y2)
```

```python
plt.scatter(X, Y)

plt.plot(X1, Y2)

plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.grid()

legends = ['Датасет', 'Аппроксимация']
plt.legend(legends)
plt.show()
```
