```python
from data_from_previous_practice import get_data, LEGENDS

import numpy as np
import matplotlib.pyplot as plt

from IPython.display import clear_output

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

```python
def draw_graph(clf):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z)
    scatter = plt.scatter(data[:, 0], data[:, 1], c=ident, edgecolors="black", s=70)
    handles = scatter.legend_elements()[0]
    plt.legend(handles, LEGENDS)
    plt.show()
```

```python
group_1, group_2, group_3, group_4, xx, yy, Z, data, ident, x_train, x_test, y_train, y_test = get_data()
```

```python
plt.scatter(group_1[:, 0], group_1[:, 1])
plt.scatter(group_2[:, 0], group_2[:, 1])
plt.scatter(group_3[:, 0], group_3[:, 1])
plt.scatter(group_4[:, 0], group_4[:, 1])
plt.title('Данные из прошлой работы')
plt.legend(LEGENDS)
plt.grid()
plt.show()
```

# kNN

```python
def chose_best_n():
    neighbors = np.arange(3, 12)
    accuracy = np.empty(len(neighbors))
    max_value = -1
    max_neighbors = neighbors[0]

    for i, k in enumerate(neighbors):
        knn = KNeighborsClassifier(k)
        knn.fit(x_train, y_train)
        acc = 100 * knn.score(x_test, y_test)
        if acc > max_value:
            max_value = acc
            max_neighbors = neighbors[i]
        accuracy[i] = acc

    plt.plot(neighbors, accuracy, label='Кривая точности')
    plt.legend()
    plt.xlabel('Число соседей')
    plt.ylabel('Точность, %')
    plt.grid()
    plt.show()

    return max_neighbors
```

```python
k_neighbors = chose_best_n()
clf = KNeighborsClassifier(k_neighbors).fit(x_train, y_train)

result = clf.predict(x_test)

print(f'При k={k_neighbors} Точность: {accuracy_score(y_test, result):.3f}')
```

```python
draw_graph(clf)
```

# SVC

```python
def chose_core():
    cores = ['linear', 'poly', 'rbf']
    accuracy = -1
    choice = 0
    for i, core in enumerate(cores):
        clf = SVC(kernel=core)
        clf.fit(x_train, y_train)
        acc = accuracy_score(y_test, clf.predict(x_test))
        print('Точность ядра', core, '- {0:.3f}'.format(acc))
        if acc > accuracy:
            accuracy = acc
            choice = i

    return cores[choice]
```

```python
core = chose_core()
print('Выбрано ядро: ', core)
clf = SVC(kernel=core).fit(x_train, y_train)
result = clf.predict(x_test)
print('Точность: {0:.3f}'.format(accuracy_score(y_test, result)))
```

```python
draw_graph(clf)
```

# Decision tree

```python
def chose_max_depth(depth=4):
    accuracy = -1
    best_depth = 0
    for d in range(1, depth + 1):
        clf = DecisionTreeClassifier(max_depth=d)
        clf.fit(x_train, y_train)
        acc = accuracy_score(y_test, clf.predict(x_test))
        print(f'Точность с глубиной {d} - {acc:.3f}')
        if acc > accuracy:
            accuracy = acc
            best_depth = d
    print(f'Выбрана точность {best_depth}')
    return best_depth
```

```python
clf = DecisionTreeClassifier(max_depth=chose_max_depth()).fit(x_train, y_train)
draw_graph(clf)
```

```python
fig, axes = plt.subplots(dpi=350)
plot_tree(clf, feature_names=LEGENDS, class_names=LEGENDS, filled=True)
```

# Naive Bayes

```python

classifiers = {
    'kNN': KNeighborsClassifier(n_neighbors=chose_best_n()),
    'SVC': SVC(kernel=chose_core()),
    'Decision tree': DecisionTreeClassifier(max_depth=chose_max_depth()),
    'Bayes': GaussianNB(),
}
clear_output(wait=True)
True
```

```python
clf = GaussianNB().fit(x_train, y_train)
print(f'Точность баейсовского равна {accuracy_score(y_test, clf.predict(x_test)):.3f}')
draw_graph(clf)
```

```python
for name, clf in classifiers.items():
    clf.fit(x_train, y_train)
    print(f'Точность {name} = {accuracy_score(y_test, clf.predict(x_test)):.3f}')
```
